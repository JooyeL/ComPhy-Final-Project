```python
import matplotlib.pyplot as plt
import pandas as pd
df= pd.read_csv('data.txt', sep='\t')
df.columns=['age','temp_var','CO2','dust']
df

#temp_var는 측정 당시 표면 온도 -55.5도씨를 기준
#측정당시 1999년 CO2 농도: 280ppm, 
#현재: 대략 414ppm, 기온 :-40도씨
#-55-(-40)=-15 
```

```python
t=df['age']
x=df['temp_var']
y=df['CO2']
z=df['dust']
```

```python
df.plot.scatter(x='temp_var', y='CO2') 
#CO2농도가 낮을수록 온도가 낮고 높을수록 같이 높아짐
```

```python
temp_graph=df.plot.scatter(x='age',y='temp_var', color='red', figsize=(10,8))
```

```python
CO2_graph=df.plot.scatter(x='age',y='CO2',figsize=(10,8)) 
#위 두 그래프가 거의 일치하며 온도변화와 CO2농도의 관계성을 보여줌
```

```python
# 데이터의 평균값 계산
x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)

# 기울기와 y 절편 계산
numerator = sum((xi - x_mean) * (yi - y_mean) for xi, yi in zip(x, y))
denominator = sum((xi - x_mean) ** 2 for xi in x)

a_fit = numerator / denominator  # 기울기
b_fit = y_mean - (a_fit * x_mean)  # y 절편

# 피팅된 함수와 원본 데이터 그래프로 시각화
plt.scatter(x, y)
plt.plot(x, [a_fit * xi + b_fit for xi in x], 'r-')
plt.xlabel('Temperature_var')
plt.ylabel('CO2')

print('slope= ', a_fit,', ', 'y-intercept= ', b_fit)

```

```python
a=a_fit
b=b_fit
y= (414-b)/a
y

#현재 CO2농도 414ppm으로 예측해본 온도 변화
```
